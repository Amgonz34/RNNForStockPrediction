```python
#import needed packages
import pandas as pd

import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("ArticalData.csv")
```

```python
Time = df["Date"]
```

```python
df = df.fillna("None")
```

```python
Sources = df["Source"]
```

```python
#create Dummy values for Source and Newsdesk attributes
DummySources = pd.get_dummies(Sources, columns=['Source'])
DummyNewsDesk = pd.get_dummies(df["News Desk"], columns=['News Desk'])
```

```python
#create Dummy values for Artical Subjects and Artical Key Words
```

```python
count = CountVectorizer(binary = False)
bow = count.fit_transform(df['Subjects'].values)

DummySubjects = pd.DataFrame(bow.toarray(), columns = count.get_feature_names())
```

```python
count = CountVectorizer(binary = True)
bow = count.fit_transform(df['KeyWords'].values)

DummyKeyWords = pd.DataFrame(bow.toarray(), columns = count.get_feature_names())

```

Loops to iterate through the columns to only keep Dummy values that meet a minimum number of instances with in the data set.

this is set to 50 in this case

```python
temp = []
for (columnName, columnData) in DummyKeyWords.iteritems():
    sums = DummyKeyWords[columnName].sum()
    if(sums > 50):
        temp.append(columnName)
```

```python
temp2 = []
for (columnName, columnData) in DummySubjects.iteritems():
    sums = DummySubjects[columnName].sum()
    if(sums > 50):
        temp2.append(columnName)
```

Make new data frames that only references the columns selected from the above loops

```python
DummyKeyWords1 = DummyKeyWords[temp]
```

```python
DummySubjects1 = DummySubjects[temp2]
```

merge data frames

```python
DummyVariables = pd.concat([ Time ,DummySources, DummyNewsDesk, DummySubjects1,DummyKeyWords1], axis=1, sort=False)
```

```python
#save data frame
DummyVariables.to_csv('DummyVariables.csv')
```
